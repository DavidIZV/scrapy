## Lectura de todos los ficheros del directorio

```python
from os import walk
import pandas as pd

path = './../generated/ml/preprocess'
dataframes = []
for (dirpath, dirnames, filenames) in walk(path):
    for filename in filenames:
        relative_path = path + '/' + filename
        datas = pd.read_csv(relative_path, encoding = 'utf-8')
        dataframes.append(datas)
    break
dataframes

all = dataframes[0]
for i in range(1, len(dataframes)):
    all = pd.concat([all, dataframes[i]])

print(len(all))
```

## Escritura y lectura del nuevo fichero unificado

```python
all.to_csv('./../generated/ml/preprocess/all.csv', index=False)

original_data = pd.read_csv('./../generated/ml/preprocess/all.csv')
original_data
```
